```python
import csv
import pandas as pd 
import random 
import copy
```

```python
def cut_random_oligo(sequence, cut_size):
    valid_cut_site = len(sequence) - cut_size
    cut_position = random.randint(0, valid_cut_site)
    return sequence[cut_position : cut_position + cut_size]

def cut_random_oligo_in_range(sequence, bottom_range, top_range):
    r = random.randint(bottom_range, top_range)
    valid_cut_site = len(sequence) - r
    cut_position = random.randint(0, valid_cut_site)
    return sequence[cut_position : cut_position + r]
```

```python
# takes in the list of lists from the input csv file without the column row
# returns a list of lists to be turned into a csv file
def add_changes_heysol_requested(csv_lines):
    #adding 'G' to beginning of the "Non-targeting gRNA from Brunello" column 
    csv_lines2 = copy.deepcopy(csv_lines)
    for row in csv_lines2:
        if row[1][0] != "G":
            row[1] = "G" + row[1]
        #concatenating the 4 different hmr columns 
        hmr = row[5] + row[7] + row[9] + row[11]
        pbs = row[3]
        cut_hmr = cut_random_oligo_in_range(hmr, 10, 30)
        cut_pbs = cut_random_oligo_in_range(pbs, 10, 15)
        row.append(cut_pbs)#, cut_hmr])
        row.append(cut_hmr)
    return csv_lines2#[0], csv_lines2[1]       
```

```python
lines = []
with open("Brunello_nontargeting_gRNAs - PEG_ctrl.csv", "r") as f:
    contents = f.read().split('\n')
    for c in contents:
        lines.append(c.split(','))

columns = lines[0]
#print(columns)
file_rows = lines[1:]
new_csv_lines = add_changes_heysol_requested(file_rows)
print(new_csv_lines)#[0][12][0])
#new=[new_csv_lines[:][1],new_csv_lines[0][2],new_csv_lines[0][3],new_csv_lines[0][4],new_csv_lines[0][5],new_csv_lines[0][6],new_csv_lines[0][7],new_csv_lines[0][8],new_csv_lines[0][9],new_csv_lines[0][10],new_csv_lines[0][11],new_csv_lines[0][12][0],new_csv_lines[0][12][1]]   
#new=new_csv_lines[1],new_csv_lines[2],new_csv_lines[3],new_csv_lines[4],new_csv_lines[5],new_csv_lines[6],new_csv_lines[7],new_csv_lines[8],new_csv_lines[9],new_csv_lines[10],new_csv_lines[11],new_csv_lines[12][0],new_csv_lines[12][1]
#print(new)
output_file = "james_control_edits_sat.csv"

new_columns = columns + ["pbs_cut_to_10_15_nt", "HMR_cut_to_10_30_nt"]
new_csv_writer = [new_columns] + new_csv_lines
#print(new_csv_writer)

with open(output_file, "w") as f2:
    csv.writer(f2).writerows(new_csv_writer)
    
##############################################
#
#
# output file is james_control_edits.csv
# running this cell should do everything 
# Let me know if i fucked it up -- James
#
#
##############################################
```
